## cleaning unwanted strings from the data 

```python
df['episodes']=df["episodes"].astype('string')
df['overview']=df['overview'].astype('string')
df['duration']=df['duration'].str.replace('per ep.','')
df['duration']=df['duration'].str.replace('  ','')
df['aired_time']=df['aired_time'].str.replace(', ','/')
```

```python
df['Genres']=df['Genres'].str.replace('[','')
df['Genres']=df['Genres'].str.replace(']','')
df['Genres']=df['Genres'].str.replace("'",'')
df['Theme(s)']=df['Theme(s)'].str.replace(']','')
df['Theme(s)']=df['Theme(s)'].str.replace('[','')
df['Theme(s)']=df['Theme(s)'].str.replace("'",'')
df['Demographic']=df['Demographic'].str.replace('[','')
df['Demographic']=df['Demographic'].str.replace(']','')
df['Demographic']=df['Demographic'].str.replace("'",'')
df['Favorites']=df['Favorites'].str.replace("]",'')
df['Favorites']=df['Favorites'].str.replace("[",'')
df['Favorites']=df['Favorites'].str.replace("'",'')
df['Score_Rank']=df['Score_Rank'].str.replace('#','')
df['Popularity_Rank']=df['Popularity_Rank'].str.replace('#','')
```

```python
df['overview']=df['overview'].str.replace('Written by MAL Rewrite','')
df['overview']=df['overview'].str.replace('\n','')
df['overview']=df['overview'].str.replace('\'s','')
df['overview']=df['overview'].str.replace('[[]\]','')
```

```python
df['overview'][1]
```

```python
df['related_animes']=df['related_animes'].str.replace('\n','')
```

```python
df.head()
```

```python
df['platform']=df['platform'].str.replace('unknown','No information provided by MAL')
df['rating']=df['rating'].str.replace('None','No information provided by MAL')
df['platform']=df.platform.astype('category')
df['rating']=df.rating.astype('category')
df['Demographic']=df.Demographic.astype('category')
```

# extracting Characters & Related Anime information

```python
df['caracter_name']=df['caracter_name'].map(lambda x: ast.literal_eval(x))
```

```python
df['hero'] = df.apply(lambda row: [name for name, value in row['caracter_name'].items() if value == 'Main'], axis=1)    
```

```python
len(df['hero'])
```

```python
hero= []
for i in range(len(df['hero'])):
    if len(df['hero'][i]) > 0:
        hero.append(df['hero'][i][0])
    else:
        hero.append(np.nan)
```

```python
df['hero']=hero
df['hero']
```

```python
df['sup_heros'] = df.apply(lambda row: [name for name, value in row['caracter_name'].items() if value == 'Main'], axis=1)    
```

```python
sup_heros= []
for i in range(len(df['sup_heros'])):
    if len(df['sup_heros'][i]) > 0:
        sup_heros.append(df['sup_heros'][i][1:])
    else:
        sup_heros.append(np.nan)
```

```python
df['sup_heros']=sup_heros
```

```python
df.head()
```

```python
df['supporters'] = df.apply(lambda row: [name for name, value in row['caracter_name'].items() if value == 'Supporting'], axis=1)    
```

```python
df['related_animes']=df['related_animes'].map(lambda x: ast.literal_eval(x))
```

```python
Adaptation=[]
Prequel=[]
Alternative_version=[]
Side_story=[]
Sequel=[]
for i in range(len(df)):
    Adaptation.append(df['related_animes'][i].get('Adaptation'))
    Prequel.append(df['related_animes'][i].get('Prequel'))
    Alternative_version.append(df['related_animes'][i].get('Alternative version'))
    Side_story.append(df['related_animes'][i].get('Side story'))
    Sequel.append(df['related_animes'][i].get('Sequel'))
```

```python
df['Adaptation']=Adaptation
df['Prequel']=Prequel
df['Alternative_version']=Alternative_version
df['Side_story']=Side_story
df['Sequel']=Sequel
df[['Name','Adaptation','Prequel','Alternative_version','Side_story','Sequel']]
```
